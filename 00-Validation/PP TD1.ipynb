```python
import csv
```

```python
X = []
Y = []
with open('filtrage_colonne/test.out', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:
        try:
            X.append(int(row[0]))
            Y.append(float(row[1]))
        except:
            continue
    
```

```python
X1 = []
Y1 = []
with open('filtrage_ligne/test.out', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:
        try:
            X1.append(int(row[0]))
            Y1.append(float(row[1]))
        except:
            continue
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18, 16), dpi= 100)
plt.semilogy(X,Y, 'r', X1, Y1, 'b')
plt.show()
```
