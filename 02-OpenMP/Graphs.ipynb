```python
import matplotlib.pyplot as plt

def parse_data(src):
    a = open(src).readlines()[5:]
    return [int(x.strip().split(" ")[0]) for x in a], [float(x.strip().split(" ")[1]) for x in a]
```

```python
a = parse_data("exo2-moyenne2vecteurs/results.fast.txt")
b = parse_data("exo2-moyenne2vecteurs/results.txt")
plt.title("Average of 2 vectors")
plt.xlabel("Number of elements in vector")
plt.ylabel("Parallel time / Sequential time")
plt.semilogx(a[0], a[1], 'pink', label="8 threads") # 4 threads
plt.semilogx(b[0], b[1], 'purple', label="2 threads") # 2 threads
plt.legend()
plt.savefig("avg_vec")
plt.show()
```

```python
a = parse_data("exo3-produit-scalaire/results.fast.txt")
b = parse_data("exo3-produit-scalaire/results.txt")
plt.title("Dot product")
plt.xlabel("Number of elements in vector")
plt.ylabel("Parallel time / Sequential time")
plt.semilogx(a[0], a[1], 'pink', label="8 threads") # 4 threads
plt.semilogx(b[0], b[1], 'purple', label="2 threads") # 2 threads
plt.legend()
plt.savefig("dot_product")
plt.show()
```
