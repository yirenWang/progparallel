```python
dpfile = "results_dp.txt"

res = [x.strip().split('\t') for x in open(dpfile).readlines()]
```

```python
gpu = {}
res_gpu = res[1:-2]
for i in range(len(res_gpu)):
    if i%2 == 0:
        k = res_gpu[i][0].split(' = ')[1]
        gpu[int(k)] = float(res_gpu[i+1][1])
```

```python
import matplotlib.pyplot as plt
```
